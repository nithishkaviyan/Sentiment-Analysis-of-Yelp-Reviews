```python
import pandas as pd
import json
```

```python
#Path for json file
path=open("yelp_academic_dataset_user.json",'r',encoding='utf8')
```

```python
##Read json file
data_list=[]
for i,x in enumerate(path):
    data_list.append(json.loads(x))
```

```python
##Convert list to pandas DataFrame
data_users=pd.DataFrame.from_dict(data_list,orient='columns')
```

```python
col_names=['user_id', 'name', 'yelping_since', 'review_count', 'average_stars', 'friends', 'fans', 'useful',
           'compliment_cool', 'compliment_cute', 'compliment_funny', 'compliment_hot', 'compliment_list',
           'compliment_more', 'compliment_note', 'compliment_photos', 'compliment_plain', 'compliment_profile',
           'compliment_writer', 'cool', 'elite', 'funny' ]  
```

```python
data_users=data_users[col_names]
```

```python
data_users
```

```python
##Frequency Bar Graph for number of reviews by users
bins=[0,20,50,100,300,500,750,1000,2500,13000]
data_users['Range']=pd.cut(data_users.review_count,bins,include_lowest=True)
barplot_data=data_users[['Range','review_count']].groupby('Range').count()
barplot_data.columns=['Review Count']
barplot_data
```

```python
barplot_data.plot(kind='bar',title='Frequency plot of Review Counts').set_ylabel('Number of users')
```

```python
data_select_user=data_users[data_users.review_count>1000][['user_id','review_count']].reset_index(drop=True)
```

```python
data_select_user.to_csv('select_users.csv',sep=',',encoding='utf-8')
```

```python
data_select_users_neg=data_users[(data_users.review_count>50) & (data_users.review_count<1000)][['user_id','review_count']]\
            .reset_index(drop=True)
```

```python
data_select_users_neg.to_csv('select_users_neg.csv',sep=',',encoding='utf-8')
```
